## You Try

Merge the following data sets using all for ways: inner, left, right, and outer. See if you can predict before running the code what the output will be!

```python
df_animals = pd.DataFrame({
    'animal_id': [1, 2, 3, 4],
    'name': ['Leo', 'Stripes', 'Spot', 'Fluffy'],
    'type': ['Lion', 'Tiger', 'Cheetah', 'Cat']
})

df_habitats = pd.DataFrame({
    'animal_id': [1, 2, 5, 4],
    'habitat': ['Savannah', 'Jungle', 'Mountains', 'Domestic'],
    'population_estimate': [25000, 3200, 120, 50000000]
})

display(df_animals)
display(df_habitats)
```

## Your Prediction
Only animal_ids that exist in both df_animals and df_habitats will appear

```python
## Your code
inner_merge = pd.merge(df_animals, df_habitats, on='animal_id', how='inner')
display(inner_merge)
```

```python
## Repeat for left, right, outer
right_merge = pd.merge(df_animals, df_habitats, on='animal_id', how='right')
display(right_merge)

left_merge = pd.merge(df_animals, df_habitats, on='animal_id', how='left')
display(left_merge)

outer_merge = pd.merge(df_animals, df_habitats, on='animal_id', how='outer')
display(outer_merge)
```
